```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("merged_tennis_files/merged_file.csv")
sns.set_theme(style="whitegrid")
```

```python
df
```

```python
pair_df = pd.concat([
    pd.DataFrame({
        'aces': df['w_ace'],
        'double_faults': df['w_df'],
        'rank': df['winner_rank'],
        'age': df['winner_age'],
        'height': df['winner_ht'],
        'role': 'Winner'
    }),
    pd.DataFrame({
        'aces': df['l_ace'],
        'double_faults': df['l_df'],
        'rank': df['loser_rank'],
        'age': df['loser_age'],
        'height': df['loser_ht'],
        'role': 'Loser'
    })
]).dropna()
```

```python
sns.pairplot(
    pair_df,
    hue='role',
    diag_kind='kde',
    corner=True,
    plot_kws={'alpha': 0.3}
)

plt.show()
```
