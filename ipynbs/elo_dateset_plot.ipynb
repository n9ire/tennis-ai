```python
import pandas as pd

df = pd.read_csv("../merged_tennis_files/tennis_matches_ml_with_elo.csv")  # replace with your file path
```

```python
print(df.head())
print(df.dtypes)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

features = [
    "player_a_age",
    "player_a_rank",
    "player_a_height",
    "player_b_age",
    "player_b_rank",
    "player_b_height",
    "rank_diff",
    "age_diff",
    "height_diff",
    "elo_diff"
]

sns.pairplot(
    df[features + ["player_a_win"]],
    hue="player_a_win",
    diag_kind="hist"
)

plt.show()
```
